```python
# import necessary libraries
import pandas as pd
```

```python
# load dataset
data = pd.read_csv("NYPD_Shooting_Incident_Data__Historic_Cleaned.csv")
data.head()
```

```python
# data backup
df = data.copy()
df.info()
```

```python
# checking data distribution
df.describe()
```

```python
# checking no of values
for i in df.columns:
    print(df[i].value_counts(), "\n")
```

```python
# map JURISDICTION_CODE
mapping = {
    0: "Patrol",
    1: "Transit",
    2: "Housing"
}
df['JURISDICTION_CODE'] = df['JURISDICTION_CODE'].map(mapping)
df['JURISDICTION_CODE'].value_counts()
```

```python
# save data
df.to_csv("NYPD_Shooting_Incident_Data__Historic_Cleaned.csv", index= False)
```
